```python
import pandas as pd
df = pd.read_excel("C:/Users/hp/Downloads/ApexPlanet_DataAnalytics_Dataset.xlsx")
df.head()
```

```python
df.info()
df.describe()
df.columns
```

```python
#FIX DATE COLUMN
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df.info()
```

```python
# Fix Age
df['Age'] = df['Age'].fillna(df['Age'].median())

# Fix City
df['City'] = df['City'].fillna('Unknown')

df.info()
```

```python
#CONVERTING AGE TO INTEGER DATA TYPE
df['Age'] = df['Age'].astype(int)
df.info()
```

```python
#CLEANING TEXT FORMATTING
df['Gender'] = df['Gender'].str.capitalize()
df['City'] = df['City'].str.title()
df['Product'] = df['Product'].str.title()
df['Category'] = df['Category'].str.title()
```

```python
#VALLIDATE SALES DATA
df['Calculated_Sales'] = df['Quantity'] * df['Unit_Price']
(df['Calculated_Sales'] - df['Total_Sales']).sum()
```

```python
#FEATURE ENGINEERING
# Age group
df['Age_Group'] = pd.cut(df['Age'],
                        bins=[0,25,40,60,100],
                        labels=['Young','Adult','Middle Age','Senior'])

# Order month
df['Order_Month'] = df['Order_Date'].dt.month

# High value orders
df['High_Value_Order'] = df['Total_Sales'].apply(
    lambda x: 1 if x > 100000 else 0
)
```

```python

# FEATURE ENGINEERING VALIDATION


print("🔹 Preview of New Columns:")
print(df[['Age', 'Age_Group', 'Order_Date', 'Order_Month', 'Total_Sales', 'High_Value_Order']].head())

print("\n🔹 Age Group Distribution:")
print(df['Age_Group'].value_counts())

print("\n🔹 Order Month Unique Values:")
print(df['Order_Month'].unique())

print("\n🔹 High Value Orders Distribution:")
print(df['High_Value_Order'].value_counts())

print("\n🔹 Age vs Age Group Check:")
print(df[['Age', 'Age_Group']].sort_values(by='Age').head(10))

print("\n🔹 High Value Orders Validation (>100000):")
print(df[df['Total_Sales'] > 100000][['Total_Sales', 'High_Value_Order']].head())

print("\n🔹 Minimum value among High Value Orders (should be >100000):")
print(df[df['High_Value_Order'] == 1]['Total_Sales'].min())
```

```python
df.to_csv("cleaned_dataset.csv", index=False)
```
